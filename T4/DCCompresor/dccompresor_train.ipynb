```python
import os
import pandas as pd

# Cambiar si se descarga en otro directorio
base_path = "20_newsgroups"

def load_data(base_path):
    data = []
    for category in os.listdir(base_path):
        category_path = os.path.join(base_path, category)

        if os.path.isdir(category_path):
            for file_name in os.listdir(category_path):
                file_path = os.path.join(category_path, file_name)
                try:
                    with open(file_path, 'r', errors='ignore') as file:
                        content = file.read()
                        data.append({"category": category, "content": content})
                except Exception as e:
                    print(f"Error al leer {file_path}: {e}")
    return pd.DataFrame(data)

data_inicial = load_data(base_path)

def generate_word_pairs(texts):
    pairs = []
    for text in texts:
        words = text.split()
        for i in range(len(words) - 1):
            pairs.append((words[i], words[i + 1]))
    return pairs

pairs = generate_word_pairs(data_inicial["content"])

df = pd.DataFrame(pairs, columns=["word", "next_word"])

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Cambiar palabras a números
word_encoder = LabelEncoder()
df["word_encoded"] = word_encoder.fit_transform(df["word"])
df["next_word_encoded"] = word_encoder.fit_transform(df["next_word"])

df.info()
```

```python
word_encoder = LabelEncoder()
df["word_encoded"] = word_encoder.fit_transform(df["word"])
df["next_word_encoded"] = word_encoder.fit_transform(df["next_word"])

# Paso 2: Preparar los datos, cambiar el tamaño de la muestra para mejorar el modelo
X = df["word_encoded"].values[:50000].reshape(-1, 1)  # Características (palabras actuales)
Y = df["next_word_encoded"].values[:50000]  # Etiquetas (próximas palabras)


# Dividir en entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # Aumentar test_size para mejorar el modelo

# Entrenar modelo
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42) # Aumentar n_estimators y max_depth para mejorar el modelo
model.fit(X_train, Y_train)

# Evaluar modelo
Y_pred = model.predict(X_test)
print(classification_report(Y_test, Y_pred))
```
